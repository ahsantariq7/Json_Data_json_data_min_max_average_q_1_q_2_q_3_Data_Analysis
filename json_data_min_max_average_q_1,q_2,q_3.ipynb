# Q.2
What were the top three highest closing days for the stock. Note that the closing price for each day is price_D: last. Also, you need to provide the date on which the highest closing happened, not the price. 

```python
# Step 2: Extract dates and corresponding last prices
closing_prices = [(obj['date'], obj['last']) for obj in data['price_D']]

# Step 3: Sort the closing prices in descending order
sorted_prices = sorted(closing_prices, key=lambda x: x[1], reverse=True)

# Step 4: Retrieve the top three highest closing dates
top_three_dates = [date for date, _ in sorted_prices[:3]]

# Step 5: Print the top three dates
for date in top_three_dates:
    print(date)
```
